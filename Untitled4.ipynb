```python
%%sql
/* 1. Create the table with STRICT types */
CREATE TABLE tech_inventory (
  item_id INTEGER PRIMARY KEY,
  item_name TEXT,
  category TEXT,
  price REAL
) STRICT;

/* 2. Insert at least 3 items (CREATE operation) */
INSERT INTO tech_inventory (item_name, category, price) VALUES
  ('Raspberry Pi 5', 'Hardware', 80.00),
  ('Arduino Uno', 'Hardware', 25.00),
  ('WiFi Radar Sensor', 'Sensors', 120.00);
```

```python
%%sql
/* Show all items that cost more than $50 */
SELECT *
FROM tech_inventory
WHERE price > 50;
```

```python
%%sql
/* Update a price */
UPDATE tech_inventory
SET price = 99.99
WHERE item_name = 'WiFi Radar Sensor';

/* Delete an item */
DELETE FROM tech_inventory
WHERE item_name = 'Arduino Uno';
```
