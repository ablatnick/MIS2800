# Answer the Questions

1. Scaling the Design: In DataCamp, we practiced Normalization. In the Redshift architecture, why is "Columnar Storage" often more important for an analyst than having perfectly normalized tables? https://docs.aws.amazon.com/redshift/latest/dg/c_columnar_storage_disk_mem_mgmnt.html

2. Speed to Insight: How does the architecture of a "Leader Node" and "Compute Nodes" change how quickly a Business Intelligence report is generated?  https://docs.aws.amazon.com/redshift/latest/dg/c_high_level_system_architecture.html


## YOUR ANSWERS TO PART 2 BELOW

Question 1: Columnar storage matters more for analysts because they usually query specific columns across millions of rows. It makes reads way faster and reduces I/O. Perfect normalization is good for organization, but speed and performance matter more when running large BI queries.

Question 2: The Leader Node manages the query and distributes the work to multiple Compute Nodes. That parallel processing means large reports get split up and processed at the same time. Instead of one machine doing everything, it’s distributed, which makes BI reports way faster.
